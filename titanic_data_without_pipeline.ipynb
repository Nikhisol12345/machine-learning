```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
```

```python
df_train = pd.read_csv('/content/train.csv')
```

```python
df_train.drop(columns=['PassengerId','Ticket','Cabin','Name'], inplace=True)
```

```python
X_train,X_test,y_train,y_test = train_test_split(df_train.drop(columns=['Survived']),df_train['Survived'],test_size=0.2,random_state=42)
```

```python
df_train.isnull().sum()
```

```python

si_age = SimpleImputer()
new_age = si_age.fit_transform(X_train[['Age']])
new_tage = si_age.fit_transform(X_test[['Age']])
si_Embarked = SimpleImputer(strategy='most_frequent')
new_Embarked = si_Embarked.fit_transform(X_train[['Embarked']])
new_tEmbarked = si_Embarked.fit_transform(X_test[['Embarked']])
```

```python
on_Embarked = OneHotEncoder(sparse=False,handle_unknown='ignore')
new_on_Embarked = on_Embarked.fit_transform(new_Embarked)
new_on_tEmbarked = on_Embarked.fit_transform(new_tEmbarked)
on_Sex = OneHotEncoder(sparse=False,handle_unknown='ignore')
new_on_sex = on_Sex .fit_transform(X_train[['Sex']])
new_on_tsex = on_Sex .fit_transform(X_test[['Sex']])
```

```python
rem_data = X_train.drop(columns=['Age','Sex','Embarked'])
```

```python
rem_test = X_test.drop(columns=['Age','Sex','Embarked'])
```

```python
final_data = np.concatenate((rem_data,new_on_Embarked,new_on_sex,new_age),axis=1 )
final_tdata = np.concatenate((rem_test,new_on_tEmbarked,new_on_tsex,new_tage ),axis=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(final_data,y_train)
```

```python
y_pred = dtc.predict(final_tdata)
```

```python
accuracy_score(y_test,y_pred)
```
