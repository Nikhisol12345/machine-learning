```python
#how to find cross val score using neg mean absolute error
```

```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/housing.data"
```

```python
names = ['CRIM','ZN','INDUS', 'CHAS', 'NOX', 'RM','AGE','DIS', 'RAD','TAX','PTRATIO','B','LSTAT','MEDV']
```

```python
data = pd.read_csv(url,delim_whitespace=True,names=names)
```

```python
data
```

```python
X = data.iloc[:,0:13].values
```

```python
y = data.iloc[:,13].values
```

```python
y
```

```python
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
```

```python
seed =7
```

```python
k_fold = model_selection.KFold(n_splits= 10,shuffle=True,random_state=seed)
```

```python
scoring = 'neg_mean_absolute_error'
```

```python
model = LinearRegression()
```

```python
result = model_selection.cross_val_score(model,X,y,cv= k_fold, scoring= scoring)
```

```python
result.mean()
result.std()
```
