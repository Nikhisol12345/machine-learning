```python
import numpy as np
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as lda
```

```python
X = np.array([[-1,-1],
              [-2,-1],
              [-3,-2],
              [1,1],
              [2,1],
              [3,2]])
```

```python
y = np.array([[1,1,1,2,2,2]])
y=y.reshape(-1,1)
```

```python
LDA = lda()
```

```python
LDA.fit(X,y)
```

```python
LDA.predict([[1,2]])
```
