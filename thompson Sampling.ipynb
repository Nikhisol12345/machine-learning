```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
ds = pd.read_csv('/content/Ads_CTR_Optimisation.csv')
ds.head()
```

```python
d = 10
numbers_of_reward_1 = [0]*d
numbers_of_reward_0 = [0]*d
N = 10000
ads_selection = []
total_selection = [0]*d
total_reward = 0
numbers_of_selection = [0]*d
for n in range(0,N):
  max_random =0 
  ad = 0
  
  for i in range(0,d):
    random_beta = random.betavariate(numbers_of_reward_1[i] + 1,numbers_of_reward_0[i] + 1) #random.betavariate(alpa,beta) is use for calculatind probablity approach
    if random_beta > max_random:
      max_random = random_beta
      ad = i
  ads_selection.append(ad)
  reward = ds.values[n,ad]
  total_reward += reward
  if reward ==1:
    numbers_of_reward_1[ad] = numbers_of_reward_1[ad] +1
  else:
    numbers_of_reward_0[ad] = numbers_of_reward_0[ad] + 1

  numbers_of_selection[ad ] = numbers_of_selection[ad] +1 
```

```python
plt.bar(range(1,11),numbers_of_selection)
plt.xticks(range(1,11))
plt.show()
```
