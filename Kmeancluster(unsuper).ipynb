```python
import pandas as pd
```

```python
from sklearn.cluster import KMeans
import numpy as np
```

```python
X = np.array([
              [1,2],
              [2,4],
              [1,2],
              [10,4],
              [11,4],
              [10,5]

])
```

```python
k_mean = KMeans(n_clusters=2,random_state=0)
```

```python
k_mean.fit(X)
```

```python
y_pred = k_mean.predict([[1,3]])
```

```python
y_pred
```

```python
k_mean.cluster_centers_
```
