```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/New Text Document (4).csv')
```

```python
df
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns = ['has_covid']),df['has_covid'],test_size=0.2)
```

```python
y_train


```

```python
simple = SimpleImputer()
X_train_fever = simple.fit_transform(X_train[['fever']])

X_test_fever = simple.fit_transform(X_test[['fever']])
X_train_fever.shape
```

```python
on = OneHotEncoder(drop='first',sparse=False)
X_train_city_gender = on.fit_transform(X_train[['city','gender']])
X_test_city_gender = on.fit_transform(X_test[['city','gender']])
X_train_city_gender.shape
```

```python
X_train_city_gender
```

```python
od = OrdinalEncoder(categories=[['Mild','Strong']])
X_train_cough = od.fit_transform(X_train[['cough']])
X_test_cough = od.fit_transform(X_test[['cough']])
X_train_cough.shape
```

```python
X_train_age = X_train.drop(columns = ['cough','gender','fever','city']).values
X_test_age = X_test.drop(columns = ['cough','gender','fever','city']).values
X_train_age.shape
```

```python
X_train_transformed = np.concatenate((X_train_age,X_train_fever,X_train_city_gender,X_train_cough),axis=1)
# also the test data
X_test_transformed = np.concatenate((X_test_age,X_test_fever,X_train_city_gender,X_test_cough),axis=1)

X_train_transformed.shape
```

```python
from sklearn.compose import ColumnTransformer
```

```python

cd = ColumnTransformer(transformers = [
    ('taf1',SimpleImputer(),['fever']),
    ('taf2',OneHotEncoder(drop='first',sparse=False),['city','gender']),
    ('taf3',OrdinalEncoder(categories=[['Mild','Strong']]),['cough'])],remainder='passthrough'
)
```

```python
X_train_transformed = cd.fit_transform(X_train)
X_test_transformed =  cd.fit_transform(X_test)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='sigmoid', gamma=1.0)
```

```python
le = LabelEncoder()
y_train1 = le.fit_transform(y_train)
y_train1
```

```python
svc.fit(X_train_transformed,y_train)
```

```python
y_pred = svc.predict(X_test_transformed)
```
