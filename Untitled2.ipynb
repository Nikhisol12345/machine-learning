```python
import pandas as pd
```

```python
dataset = pd.read_csv("/content/diabetes.csv")
```

```python
dataset
```

```python
X = dataset.iloc[:,[1,2]].values
```

```python
X
```

```python
y = dataset.iloc[:,8].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_trian,X_test,y_trian,y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
X_trian
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_trian = sc.fit_transform(X_trian)
```

```python
X_test = sc.fit_transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtcf = DecisionTreeClassifier(splitter="random",random_state=0)
```

```python
dtcf.fit(X_trian,y_trian)
```

```python
y_pred = dtcf.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("accuracy ==",accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test,y_pred))
```
