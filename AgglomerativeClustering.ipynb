```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('/content/Mall_Customers.csv')
```

```python
data.head()
```

```python
X = data.iloc[:,[3,4]].values
```

```python
X
```

```python
import scipy.cluster.hierarchy as sch
```

```python
dg = sch.dendrogram(sch.linkage(X,method='ward'))
plt.title("dendrogram")
plt.xlabel("annual income")
plt.ylabel("spending score")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
agg = AgglomerativeClustering(n_clusters=5)
```

```python
predict_model = agg.fit_predict(X)
```

```python
predict_model
```

```python
plt.scatter(X[predict_model == 0,0],X[predict_model == 0,1],c="black",label = 'label1')
plt.scatter(X[predict_model == 1,0],X[predict_model == 1,1],c="red",label = 'label2')
plt.scatter(X[predict_model == 2,0],X[predict_model == 2,1],c="green",label = 'label3')
plt.scatter(X[predict_model == 3,0],X[predict_model == 3,1],c = "blue",label = 'label4')
plt.scatter(X[predict_model == 4,0],X[predict_model == 4,1],c = "pink",label = 'label5')
plt.legend()
plt.title("agglomerative")
plt.xlabel("annual income")
plt.ylabel("montly spend")
plt.show()
```
