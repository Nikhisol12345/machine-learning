```python
import pandas as pd
```

```python
data = pd.read_csv('/content/Market_Basket_Optimisation.csv')
```

```python
transation = []
```

```python
for i in range(0,7500):
  transation.append([str(data.values[i,j]) for j in range(0,20)])
```

```python
transation
```

```python
!pip install apyori
```

```python
from apyori import apriori
```

```python
rule = apriori(transation,min_support = 0.01,min_confidence = 0.2,min_lift = 0.3,max_length = 3)
```

```python
result = list(rule)
```

```python
result_list = []
```

```python
result
```

```python
for i in range(0,len(result)):
  result_list.append('rule:'+ str(result[i][0]) + 'supportt' + str(result[i][1]))
```
