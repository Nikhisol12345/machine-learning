```python

import numpy as np
import mlxtend.frequent_patterns as fp
```

```python

data = [['milk','onion','beans'],
        ['milk','onion'],
        ['onion','egg','yogurt']]
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
```

```python
model = te.fit_transform(data)
```

```python
model
```

```python
dataset = pd.DataFrame(model,columns=te.columns_)
```

```python
dataset
```

```python
fp(dataset,min_support = 0.60)
```
