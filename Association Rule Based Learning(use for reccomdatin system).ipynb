```python
import pandas as pd
```

```python
data = [['Milk','Onion','Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],

['Dill','Onion', 'Nutmeg', 'Kidney Beans'],

['Milk', 'Apple', 'Kidney Beans', 'Eggs'],

['Milk', 'Unicorn', 'Corn','Yogurt'],

['Corn', 'Onion', 'Onion', 'Kidney','Yogurt'],

['Kidney Beans', 'Beans', 'Ice cream', 'Eggs']]

```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
```

```python
dataset = te.fit(data).transform(data)
```

```python
finaldata = pd.DataFrame(dataset,columns=te.columns_)
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
apriori(finaldata,min_support= 0.4,use_colnames=True)
```
