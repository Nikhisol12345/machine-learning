```python
import numpy as np
from sklearn.decomposition import PCA
```

```python
X = np.array([[-1,-1],
              [-2,-1],
              [-3,-2],
              [1,1],
              [2,1],
              [3,2]])
```

```python
pca = PCA(n_components= 2)
```

```python
pca.fit(X) 
```

```python
pca.explained_variance_
```

```python
pca.singular_values_
```
